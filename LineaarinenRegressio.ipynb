```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('http://taanila.fi/winequality-red.csv',sep=';')

df
```

Punaviini

```python
#Datan valmistelu.

X=df['fixed acidity'].to_frame() #feature-matriisi X

y=df['quality'] #target eli selitettävä muuttuja eli y-muuttuja
```

```python
#Tuon lineaarisen mallin:
from sklearn.linear_model import LinearRegression

#Sovitan mallin dataan:
malli = LinearRegression()
malli.fit(X,y)
```

```python
malli.coef_ #kulmakerroin
```

```python
malli.intercept_ #vakiotermi
```

```python


malli.score(X,y) #selityskerroin

```

```python
#Mallin sopivuuden arviointi sea
sns.jointplot(data=df, x='fixed acidity', y='quality', kind='reg')
```
