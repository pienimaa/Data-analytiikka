```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('http://taanila.fi/winequality-red.csv',sep=';')

df
```

Punaviini

```python
sns.pairplot(df, kind='reg')
```

```python
#Riippuvuuksien tarkastelua korrelaatiokertoimien avulla:
correlation_matrix = df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
#X=features, y=target
X=df[['fixed acidity', 'sulphates', 'alcohol']]
y=df['quality']
```

```python
from sklearn.linear_model import LinearRegression

malli=LinearRegression()
malli.fit(X,y)
```

```python
malli.intercept_ 

```

```python
malli.coef_
```

```python
malli.score(X,y) #Mallin selityskerroin
```

```python
#Mallin virhetermit:
plt.scatter(malli.predict(X), malli.predict(X)-y)

#Vaakaviiva nollapoikkeaman kohdalle:
plt.hlines(y=0,xmin=2.5,xmax=10)

plt.xlabel('Ennuste')
plt.ylabel('Poikkeama todellisesta')
```

```python


#Todellisen hinnan ja ennustamallin antamien ennusteiden vastaavuus:
plt.scatter(df['quality'], malli.predict(X))

plt.xlabel('Quality')
plt.ylabel('Predicted value')

```
